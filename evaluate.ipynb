```python
from kervis.utils import Evaluator
from kervis.utils import Model
from kervis.kernels import ShortestPath
```

```python
SP = Model('MUTAG', ShortestPath, 'SVM', test_size=0.4)
```

```python
evaluator = Evaluator(SP)
```

```python
evaluator.classification_report()
```

```python
evaluator.roc_curve()
```

```python
evaluator.precision_recall_curve()
```

```python
evaluator.confusion_matrix()
```
