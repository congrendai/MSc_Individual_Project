```python
from kervis.utils import Model
from kervis.utils import Dataset
from kervis.kernels import VertexHistogram, EdgeHistogram, ShortestPath, GraphletSampling, WeisfeilerLehman
```

```python
VH = VertexHistogram()
EH = EdgeHistogram()
SP = ShortestPath()
GL = GraphletSampling(k=5)
WL = WeisfeilerLehman()
```

```python
MUTAG = Dataset("MUTAG", cmap="tab20b")
```

```python
MUTAG.plot_G(100)
```

```python
MUTAG.node_color_map
```

```python
MUTAG.plot_class_distribution()
```

```python
VH_MUTAG = Model(VH, MUTAG, 'svm')
EH_MUTAG = Model(EH, MUTAG, 'svm')
SP_MUTAG = Model(SP, MUTAG, 'svm')
GL_MUTAG = Model(GL, MUTAG, 'svm')
WL_MUTAG = Model(WL, MUTAG, 'svm')
```
