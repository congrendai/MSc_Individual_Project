```python
from kervis.utils import Model
from kervis.utils import Dataset
from kervis.kernels import VertexHistogram, EdgeHistogram, ShortestPath, Graphlet, WeisfeilerLehman
```

```python
VH = VertexHistogram()
EH = EdgeHistogram()
SP = ShortestPath()
GL = Graphlet()

MUTAG = Dataset("MUTAG")
```

```python
VH_MUTAG = Model(VH, MUTAG, "SVM")
EH_MUTAG = Model(EH, MUTAG, "SVM")
SP_MUTAG = Model(SP, MUTAG, "SVM")
GL_MUTAG = Model(GL, MUTAG, "SVM")
```

```python
VH_MUTAG.explain()
EH_MUTAG.explain()
SP_MUTAG.explain()
GL_MUTAG.explain()
```
